```python
import os
import requests
import time
import pandas as pd
from bs4 import BeautifulSoup


# Create output folder
output_folder = "text_dataset"
os.makedirs(output_folder, exist_ok=True)

# Function to scrape text from a webpage
def scrape_text(url):
    try:
        response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"}, timeout=10)
        if response.status_code != 200:
            return None
        
        soup = BeautifulSoup(response.text, "html.parser")
        
        # Extract text from <p> tags
        paragraphs = soup.find_all("p")
        text = "\n".join([p.get_text() for p in paragraphs])
        
        return text.strip()
    
    except Exception as e:
        print(f"Error scraping {url}: {e}")
        return None

# Scrape and save text for each category
for category, urls in websites.items():
    print(f"Scraping category: {category}")
    
    category_text = []
    for url in urls:
        print(f"Fetching: {url}")
        text = scrape_text(url)
        if text:
            category_text.append(text)
        time.sleep(2)  # Pause to avoid being blocked
    
    # Save text data to a file
    if category_text:
        file_path = os.path.join(output_folder, f"{category.replace(' ', '_')}.txt")
        with open(file_path, "w", encoding="utf-8") as file:
            file.write("\n\n".join(category_text))
    
    print(f"Saved {category} data!\n")

print("Text scraping completed!")
```

```python
import os
import re
import nltk
from nltk.corpus import stopwords

# Download stopwords (only needed once)
nltk.download("stopwords")

# Load English stopwords
stop_words = set(stopwords.words("english"))

# Define input/output folders
input_folder = "text_dataset"
output_folder = "cleaned_text_dataset"
os.makedirs(output_folder, exist_ok=True)

# Function to clean text
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'<[^>]+>', '', text)  # Remove HTML tags
    text = re.sub(r'[^a-z\s]', '', text)  # Remove punctuation and numbers
    words = text.split()
    words = [word for word in words if word not in stop_words]  # Remove stopwords
    return " ".join(words)

# Process each text file
for filename in os.listdir(input_folder):
    if filename.endswith(".txt"):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)

        with open(input_path, "r", encoding="utf-8") as file:
            text = file.read()

        cleaned_text = clean_text(text)

        with open(output_path, "w", encoding="utf-8") as file:
            file.write(cleaned_text)

        print(f"Cleaned: {filename}")

print("Text preprocessing completed!")
```

```python
import os
import time
import requests
import pandas as pd
from pydub import AudioSegment
from pydub.playback import play


# Output folder
output_folder = "audio_dataset"
os.makedirs(output_folder, exist_ok=True)

# CSV file to store metadata
csv_filename = os.path.join(output_folder, "metadata.csv")

# Function to record and save audio
def record_audio(station_name, url, duration=30):
    try:
        print(f"Recording from: {station_name} ({duration} sec)")
        response = requests.get(url, stream=True, timeout=10)
        
        if response.status_code != 200:
            print(f"Failed to access {url}")
            return None

        # Save the raw audio data
        audio_path = os.path.join(output_folder, f"{station_name.replace(' ', '_')}_{int(time.time())}.mp3")
        with open(audio_path, "wb") as audio_file:
            start_time = time.time()
            for chunk in response.iter_content(chunk_size=1024):
                if time.time() - start_time > duration:
                    break
                audio_file.write(chunk)

        # Convert to WAV format
        audio_wav_path = audio_path.replace(".mp3", ".wav")
        audio = AudioSegment.from_mp3(audio_path)
        audio.export(audio_wav_path, format="wav")

        return station_name, audio_wav_path, duration

    except Exception as e:
        print(f"Error recording {station_name}: {e}")
        return None

# Store metadata
metadata = []

for i in range(1,11):
    for station, url in radio_stations.items():
        result = record_audio(station, url, duration = 30)
        if result:
            metadata.append(result)

# Save metadata to CSV
df = pd.DataFrame(metadata, columns=["Station Name", "File Path", "Duration (sec)"])
df.to_csv(csv_filename, index=False)

print("Audio recording completed!")
```
